```python
# FILENAMES

# Foreign to California
usa_shape_filename = "data/others/stanford-vw421bb8115-geojson.json"
landuse_shape_filename = "data/others/TG95CALPYPY-geojson.json"
mexico_shape_filename = "data/others/stanford-zc863pb5331-geojson.json"
unknown_shape_filename = "data/others/c1976_2000_0/c1976_2000.shp"

# Blocks
blocksgroups_shape_filename = "data/polygons/TG00CAGRP.shp"
```

```python
# CALIFORNIA SHARED BORDERS STATES 

with open(usa_shape_filename) as json_file:
    usa_shape = json.load(json_file)
    
# OREGON
oregon_polygons = [shape(item["geometry"]) for item in usa_shape["features"] if item["properties"]["state"] == "Oregon"]
oregon_areas = np.array([item.area for item in oregon_polygons])
oregon_max_idx = np.argmax(oregon_areas)
oregon_biggest_polygon = oregon_polygons[oregon_max_idx]


# NEVADA
nevada_polygons = [shape(item["geometry"]) for item in usa_shape["features"] if item["properties"]["state"] == "Nevada"]
nevada_areas = np.array([item.area for item in nevada_polygons])
nevada_max_idx = np.argmax(nevada_areas)
nevada_biggest_polygon = nevada_polygons[nevada_max_idx]


# ARIZONA
arizona_polygons = [shape(item["geometry"]) for item in usa_shape["features"] if item["properties"]["state"] == "Arizona"]
arizona_areas = np.array([item.area for item in arizona_polygons])
arizona_max_idx = np.argmax(arizona_areas)
arizona_biggest_polygon = arizona_polygons[arizona_max_idx]


# MEXICO

with open(mexico_shape_filename) as json_file:
    mexico_shape = json.load(json_file)
    
mexico_polygon = shape(mexico_shape["features"][0]["geometry"])
areas = [item.area for item in mexico_polygon]
mexico_biggest_polygon = mexico_polygon[np.argmax(np.array(areas))]


# CALIFORNIA INNER WATERS

blocksgroups_geojson = fiona.open(blocksgroups_shape_filename)
blocks_polygons = [shape(item["geometry"]) for item in blocksgroups_geojson]

inner_water_polygons = []
for idpoly,polyexample in enumerate(blocks_polygons[:]):
    
    if polyexample.type == "Polygon":
        list_of_holes = polyexample.interiors
        
    elif polyexample.type == "MultiPolygon":
        list_of_holes = []
        for subpoly in polyexample:
            list_of_holes = list_of_holes + list(subpoly.interiors)
    else:
        list_of_holes = []
        
    temp_water_polygons = [Polygon(hole) for hole in list_of_holes]
    inner_water_polygons = inner_water_polygons + temp_water_polygons
    
california_inner_water_polygon = MultiPolygon(inner_water_polygons)
```

```python
# CALIFORNIA
california_polygons = [shape(item["geometry"]) for item in usa_shape["features"] if item["properties"]["state"] == "California"]
california_areas = np.array([item.area for item in california_polygons])
california_max_idx = np.argmax(california_areas)
california_biggest_polygon = california_polygons[california_max_idx]

# Construction d'une box enveloppant la california jusque loin au large
minx,miny,maxx,maxy = california_biggest_polygon.bounds
topright = (maxx,maxy)
bottomright = (maxx,miny)
bottomleft = (minx-10,miny)
topleft = (minx-10,maxy)
mybox = [topright,bottomright,bottomleft,topleft,topright]
mybox = Polygon(mybox)

# Construction d'un polygone représentant l'océan pacifique au large de la california
pacific_ocean_polygons = mybox.difference(california_biggest_polygon)
pacific_ocean_polygons_area = np.array([item.area for item in pacific_ocean_polygons])
pacific_ocean_polygon = pacific_ocean_polygons[np.argmax(pacific_ocean_polygons_area)]
```

```python
# SAUVEGARDE DES POLYGONES

with open('data/foreign_countries/mexico_shape.p', 'wb') as f:
    pickle.dump(mexico_biggest_polygon, f)
    
with open('./data/foreign_countries/oregon_shape.p', 'wb') as f:
    pickle.dump(oregon_biggest_polygon, f)
    
with open('data/foreign_countries/nevada_shape.p', 'wb') as f:
    pickle.dump(nevada_biggest_polygon, f)
    
with open('data/foreign_countries/arizona_shape.p', 'wb') as f:
    pickle.dump(arizona_biggest_polygon, f)
    
with open('data/foreign_countries/california_inner_waters_shape.p', 'wb') as f:
    pickle.dump(california_inner_water_polygon, f)
    
with open('data/foreign_countries/pacific_ocean_shape.p', 'wb') as f:
    pickle.dump(pacific_ocean_polygon, f)
```
